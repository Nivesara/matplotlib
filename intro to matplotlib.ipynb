# Making figures with numpy array

```python
x=np.linspace(0,10,100)
x[:100]
```

```python
#line plot
fig,ax=plt.subplots()
ax.plot(x,x**2);
```

```python
#scatter plot
fig,ax=plt.subplots()
ax.scatter(x,np.exp(x));
```

```python
# another scatter plot
fig,ax=plt.subplots()
ax.scatter(x,np.sin(x));
```

```python
#bar plot
flower ={"rose":10,
         "jasmin":18,
         "lilly":15
        }
fig,ax=plt.subplots()
ax.bar(flower.keys(),flower.values());
```

```python
fig,ax=plt.subplots()
ax.barh(list(flower.keys()),list(flower.values()));
```

```python
#hist plot
z=np.random.randn(1000)
fig,ax=plt.subplots()
ax.hist(z);
```

```python
# subplot
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,
                                      ncols=2,
                                      figsize=(10,5))
ax1.plot(x,x**2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(flower.keys(),flower.values());
ax4.hist(np.random.randn(1000));
```

```python
#subplot option2
fig,ax =plt.subplots(nrows=2,
                     ncols=2,
                     figsize=(10,5))
#plot to each other difference 
ax[0,0].plot(x,x**2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(flower.keys(),flower.values());
ax[1,1].hist(np.random.randn(1000));
```

```python
import pandas as pd
```

```python
#making dataframe
car_sales = pd.read_csv("7.1 car-sales.csv.csv")
car_sales
```

```python
ts= pd.Series(np.random.randn(1000),
             index= pd.date_range('28/6/2020',periods=1000))
ts = ts.cumsum()
ts.plot()
```

```python
car_sales["sales Date"] = pd.date_range("28/6/2020",periods=len(car_sales))
car_sales
```

```python
car_sales["Total sales"] = car_sales["Price"].cumsum()
car_sales
```

```python
type(car_sales["Price"][0])
```

```python
car_sales["Price"]=car_sales["Price"].str.replace('[\$\,\.]','')
car_sales
```

```python
car_sales["Price"]=car_sales["Price"].str[:-2]
car_sales
```

```python
car_sales["Total sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
car_sales.plot(x="sales Date",y="Total sales");
```

```python
car_sales["Price"] = car_sales["Price"].astype(int)
car_sales.plot(x="Odometer (KM)",y="Price",kind="scatter");
```

```python
heart_disease = pd.read_csv("heart-disease.csv.csv")
heart_disease
```

```python
heart_disease["age"].plot.hist(bins=50)
```

```python
heart_disease.head()
```

```python
heart_disease.plot.hist(figsize=(10,30),subplots=True);
```

### choose wise

```python
 over_50 = heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
over_50.plot(kind="scatter",
            x='age',
            y='chol',
            c='target');
```

```python
fix,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c="target",
            ax=ax);
ax.set_xlim([45,100]);
```

```python
fig,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=over_50['age'],
                  y=over_50['chol'],
                  c=over_50['target']);

ax.set(title="Heart Disease and cholesterol Levels",
      xlabel = "Age",
      ylabel="Cholestetrol");

ax.legend(*scatter.legend_elements(),title="Target");

ax.axhline(over_50["chol"].mean(),
          linestyle='--');
```

```python
fix,(ax0,ax1)=plt.subplots(nrows=2,
                          ncols=1,
                          figsize=(10,10))

scatter = ax0.scatter(x=over_50['age'],
                      y=over_50['chol'],
                     c=over_50['target'])

ax0.set(title="Heart Disease and Cholestrol Levels",
       xlabel='Age',
       ylabel='Cholestrol')

ax0.legend(*scatter.legend_elements(), title='Target')

ax0.axhline(y=over_50['chol'].mean(),
           linestyle="--");

scatter= ax1.scatter(x=over_50['age'],
                      y=over_50['thalach'],
                     c=over_50['target'])

ax0.set(title="Heart Disease and Cholestrol Levels",
       xlabel='Age',
       ylabel='Max HeartRate');

ax1.legend(*scatter.legend_elements(), title='Target')

ax1.axhline(y=over_50['thalach'].mean(),
           linestyle="--");


fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
